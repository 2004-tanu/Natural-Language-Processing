```python
text = """The year 2024 has been quite a journey! Our team completed Project Alpha, increasing efficiency by 27.5% in
just Q3. The final report (document_v3.pdf) contained over 1,240 pages of data and was shared with all stakeholders via
our secure portal: https://example.com/projectalpha. Don't worry about the 5-page summary; it covers all the essential points,
including a detailed analysis of the $1.5 million budget allocation and the impact of the 'Work-Life Balance' initiative! We need to
schedule a follow-up meeting @ 3 p.m. to discuss phase 2, focusing on our next big challenge... can't wait!"""
```

```python
# step 1 Tokenization
    # 1. word Tokenization
    # 2.sentence Tokenization
```

```python
from nltk import word_tokenize
```

```python
tokens = word_tokenize(text)

print(tokens)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(text)

print(sentences)
```

```python
# step 2 Stemming/Lemmatization
#     1.stemming
#         a.Basic Stemming
#         b.Snowball Stemmer
#     2.Lemmatization
```

```python
def basic_stem(word):
    suffixes = ["ing", "ed", "ly", "es", "s"]
    for suffix in suffixes:
        if word.endswith(suffix) and len(word) > len(suffix) + 2:
            return word[:-len(suffix)]
    return word

basic_stems = []

for word in tokens:
    basic_stems.append(basic_stem(word))

print(basic_stems[:30])
```

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

snowball_stems = []

for word in tokens:
    snowball_stems.append(stemmer.stem(word))

print(snowball_stems[:30])
```

```python
tokens = word_tokenize(text.lower())
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()

lemmas = []

for word in tokens:
    if word.isalpha():   # remove numbers & symbols
        lemmas.append(lemmatizer.lemmatize(word))

print(lemmas[:30])
```

```python
# step 3. Stopwatch Removeble
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = []

for word in tokens:
    if word.isalpha() and word not in stop_words:
        filtered_words.append(word)

print(filtered_words[:30])
```
