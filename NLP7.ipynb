## **Initial model building with Simple input**

```python
text_data = [
    "king is a strong man",
    "queen is a wise woman",
    "boy is a young man",
    "girl is a young woman",
    "prince is the son of the king",
    "princess is the daughter of the queen",
    "man and woman are adults",
    "boy and girl are children"
]
```

```python
sentences = []

for sentence in text_data:
    tokens = word_tokenize(sentence.lower())
    sentences.append(tokens)

print(sentences)
```

**1) Word2vec model**

```python
from gensim.models import Word2Vec

word2vec_model = Word2Vec(sentences=sentences,vector_size=50,window=3,min_count=1)
```

```python
word2vec_model.save("word2vec_simple.model")
```

```python
print(word2vec_model.wv['king'])
```

```python
word2vec_model.wv.most_similar('king')
```

**2) FastText model**

```python
from gensim.models import FastText

fasttext_model = FastText(sentences=sentences,vector_size=50,window=3,min_count=1)
```

```python
fasttext_model.save("fasttext_simple.model")
```

```python
print(fasttext_model.wv['kingly'])
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

**Preprocessing of dataset**

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess_review(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words]
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    return tokens
```

```python
processed_reviews = df.apply(preprocess_review)
```

```python
print(processed_reviews.iloc[0])
```

```python
sentences = processed_reviews.tolist()
```

**1) word2text model**

```python
word2vec_model = Word2Vec(sentences=sentences,vector_size=100,window=5,min_count=5,workers=4)
```

```python
word2vec_model.save("reviews_word2vec.model")
```

```python
word2vec_model.wv["battery"]
```

**Similarity of word**

```python
word2vec_model.wv.most_similar("battery")
```

```python
word2vec_model.wv.most_similar("bad")
```

```python
word2vec_model.wv.most_similar("cheap")
```

**Similarity between two words**

```python
similarity = word2vec_model.wv.similarity("battery", "charger")
print("Word2Vec similarity:", similarity)
```

```python
pairs = [
    ("battery", "charger"),
    ("phone", "case"),
    ("good", "bad"),
    ("screen", "display")
]

for w1, w2 in pairs:
    print(w1, w2,word2vec_model.wv.similarity(w1, w2))
```
