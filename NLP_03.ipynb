Statement = Viterbi Decoding Algorithm for POS: Given a sequence of words, implement the Viterbi algorithm
to     find the most likely sequence of POS tags by estimating emission and transition probabilities from a
tagged corpus.

```python
words = ["Time", "Flies"]
tags = ["N", "V"]

Pi = [0.6, 0.4]

trans = [[0.3, 0.7],
         [0.8,0.2]]

emit = [[0.5, 0.1],
        [0.2, 0.6]]

```

```python
T = len(words)
N = len(tags)

# Viterbi matrix
viterbi = [[0 for i in range(N)] for j in range(T)]

# Backpointer matrix
backpointer = [[0 for i in range(N)] for j in range(T)]

```

```python
# Initialization

for j in range(N):
   viterbi[0][j] = Pi[j] * emit[0][j]


# Recursion
for t in range(1, T):
  for j in range(N):
    max_prob = 0
    max_prob = 0

    for i in range(N):
      prob = viterbi[t-1][i] * trans[i][j] * emit[t][j]

      if prob > max_prob:
        max_prob = prob
        max_state = i

    viterbi[t][j] = max_prob
    backpointer[t][j] = max_state
```

```python
# termination

best_prob = max(viterbi[T-1])
best_tag = viterbi[T-1].index(best_prob)
```

```python
# Backtracking

result = [best_tag]
for t in range(T-1, 0, -1):
  best_tag = backpointer[t][best_tag]
  result.insert(0, best_tag)
```

```python
# Print result

print("Most likely POS tag sequence:")
for i in result:
  print(tags[i], end=" ")
```
