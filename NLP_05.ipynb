```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.wsd import lesk

text = "John went to the market. He bought apples because they were fresh."

tokens = word_tokenize(text)
pos_tags = nltk.pos_tag(tokens)

person_entity = None
object_entity = None
resolved = []

for word, tag in pos_tags:

    # Proper noun (person name)
    if tag == "NNP":
        person_entity = word
        resolved.append(word)

    # Common noun (object)
    elif tag.startswith("NN"):
        object_entity = word
        resolved.append(word)

    # Pronoun resolution
    elif tag == "PRP":
        if word.lower() in ["he", "she", "him", "her"]:
            resolved.append(person_entity)
        elif word.lower() in ["they", "them", "it"]:
            resolved.append(object_entity)
        else:
            resolved.append(word)

    else:
        resolved.append(word)

# Lesk word sense
print("Word Sense using Lesk:\n")
for word in tokens:
    sense = lesk(tokens, word)
    if sense:
        print(word, ":", sense.definition())

print("\nOriginal Text:")
print(text)

print("\nResolved Text:")
print(" ".join(resolved))
```
