```python
def min_edit_distance(str1, str2):
    m = len(str1)
    n = len(str2)

    # Create DP table
    dp = [[0 for col in range(n + 1)] for row in range(m + 1)]

    for row in range(m + 1):
        dp[row][0] = row
    for col in range(n + 1):
        dp[0][col] = col

    for row in range(1, m + 1):
        for col in range(1, n + 1):
            if str1[row - 1] == str2[col - 1]:
                dp[row][col] = dp[row - 1][col - 1]
            else:
                dp[row][col] = 1 + min(
                    dp[row - 1][col],     # Deletion
                    dp[row][col - 1],     # Insertion
                    dp[row - 1][col - 1]  # Substitution
                )

    return dp
```

```python
# Backtracking to count operations
def backtrack(dp, str1, str2):
    row = len(str1)
    col = len(str2)

    insert = 0
    delete = 0
    replace = 0

    while row > 0 and col > 0:
        if str1[row - 1] == str2[col - 1]:
            row -= 1
            col -= 1
        elif dp[row][col] == dp[row - 1][col - 1] + 1:
            replace += 1
            row -= 1
            col -= 1
        elif dp[row][col] == dp[row - 1][col] + 1:
            delete += 1
            row -= 1
        else:
            insert += 1
            col -= 1

    insert += col
    delete += row

    return insert, delete, replace
```

```python
num_words = int(input("Enter number of dictionary words: "))
dictionary = []

for index in range(num_words):
    dictionary.append(input("Enter word: "))

input_word = input("Enter misspelled word: ")


min_dist = float('inf')
best_word = ""
best_dp = None

for dict_word in dictionary:
    dp = min_edit_distance(input_word, dict_word)
    distance = dp[len(input_word)][len(dict_word)]

    if distance < min_dist:
        min_dist = distance
        best_word = dict_word
        best_dp = dp
```

```python
# Display DP Table
print("\nMinimum Edit Distance Table:")
print("    ", end="")
for ch in best_word:
    print(ch, end="  ")
print()

for row in range(len(best_dp)):
    if row == 0:
        print(" ", best_dp[row])
    else:
        print(input_word[row - 1], best_dp[row])
```

```python
# Backtracking
insertions, deletions, substitutions = backtrack(best_dp, input_word, best_word)

print("\nMisspelled Word:", input_word)
print("Corrected Word:", best_word)
print("Minimum Edit Distance:", min_dist)

print("\nOperations Performed:")
print("Insertions:", insertions)
print("Deletions:", deletions)
print("Substitutions:", substitutions)
```
