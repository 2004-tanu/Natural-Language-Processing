```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("winequality-red.csv", sep=';')
df2 = pd.read_csv("winequality-white.csv", sep=';')
# merged_df = pd.merge(df1, df2, on="id")
# print (merged_df)
```

```python
df1["id"] = range(1, len(df1) + 1)
df2["id"] = range(1, len(df2) + 1)
```

```python
merged_df = pd.merge(df1, df2, on="id", how="inner", suffixes=("_red", "_white"))
```

```python
# Separate Features & Target
```

```python
X = merged_df.drop(columns=["quality_white", "quality_red", "id"])
y = merged_df[["quality_white", "quality_red"]]
```

```python
# Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dendrogram
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10,6))
dendrogram(linked)
plt.title("Wine Quality Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
# Apply Agglomerative Clustering
```

```python
from sklearn.cluster import AgglomerativeClustering

agglo = AgglomerativeClustering(
    n_clusters=3,   # low, medium, high quality
    linkage='ward'
)

clusters = agglo.fit_predict(X_scaled)
```

```python
merged_df["cluster"] = clusters
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis', legend='full')
plt.title('Clusters of Wine Quality (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
